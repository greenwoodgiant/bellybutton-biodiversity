```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python
engine = create_engine("sqlite:///belly_button_biodiversity.sqlite")
conn = engine.connect()
```

```python
Base = declarative_base()

class OTU(Base):
    __tablename__ = 'taxonomic_units'

    otu_id = Column(Integer, primary_key=True)
    lowest_taxonomic_unit_found = Column(Text)
    
    def __repr__(self):
        return f"ID: {self.otu_id}, NAME: {self.lowest_taxonomic_unit_found}"
```

```python
Base.metadata.create_all(engine)
```

```python
to_insert = pd.read_csv("otu_clean.csv")
data = to_insert.to_dict(orient="records")
```

```python
data
```

```python
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
table = sqlalchemy.Table('taxonomic_units', metadata, autoload=True)
```

```python
conn.execute(table.insert(), data)
```
