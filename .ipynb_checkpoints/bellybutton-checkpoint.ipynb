```python
import pandas as pd
```

```python
bb1_data = pd.read_csv("bb1_csv.csv")
bb1_data.rename(columns = {"OTU ID #":"OTU ID","Lowest Taxonomic Level of Bacteria/Archaea Found":"Bacteria"},inplace=True)
bb1_data.drop(bb1_data.columns[-1],axis=1,inplace=True)
bb1_data.head()
```

```python
bb1_metadata = pd.read_csv("bb1_meta_csv.csv")
bb1_metadata.set_index(["Data"],drop=True,inplace=True)
bb1_metadata.drop(bb1_metadata.columns[-1],axis=1,inplace=True)
bb1_metadata = bb1_metadata[1:]
bb1_metadata.head()
```

```python
bb1_metadata = bb1_metadata.transpose()
bb1_metadata.reset_index(inplace=True)
bb1_metadata.columns = ['Sample ID', 'Ethnicity', 'Gender', 'Age', 'Scrubs per Week', 
                        'Innie/Outie', 'City/State', 'Country 0-12', 'Zip 0-12', 
                        'Country 13-19', 'Zip 13-19', 'Dog', 'Cat', 'Cement 0-13', 
                        'Zip Productivity 0-13', 'Max Temp 0-13', 'Forest Cover 0-13', 
                        'Cement 13-19', 'Zip Productivity 13-19', 'Max Temp 13-19', 
                        'Forest Cover 13-19']
bb1_metadata = bb1_metadata[['Sample ID','Ethnicity','Gender','Age','Scrubs per Week','Innie/Outie','City/State']]
bb1_metadata.head()
```

```python
bb1_data.to_csv("bb1_clean.csv",index=False)
bb1_metadata.to_csv("bb1m_clean.csv",index=False)
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python
engine = create_engine("sqlite:///bellybuttons.sqlite")
conn = engine.connect()
```

```python
Base = declarative_base()

class Bellybutton(Base):
    __tablename__ = 'bellybuttons'

    sample_id = Column(Integer, primary_key=True)
    ethnicity = Column(Text)
    gender = Column(Text)
    age = Column(Integer)
    scrubs_per_week = Column(Integer)
    innie_outie = Column(Text)
    city_state = Column(Text)
   
    def __repr__(self):
        return f"Sample {self.sample_id}, {self.age}yo {self.ethnicity} {self.gender} from {city_state}, innie/outie: {self.innie_outie}, {scrubs_per_week} scrubs/week"
```

```python
Base.metadata.create_all(engine)
```

```python
to_insert = pd.read_csv("bb1m_clean.csv")
data = to_insert.to_dict(orient="records")
```

```python
data
```

```python
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
table = sqlalchemy.Table('bellybuttons', metadata, autoload=True)
```

```python
conn.execute(table.delete())
conn.execute(table.insert(),data)
```
